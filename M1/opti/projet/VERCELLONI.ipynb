1. $$f(x) = \frac{1}{2}\sum_1^N (x_i - y_i)^2 + \frac{\lambda}{2}\sum_1^{N - 1} (x_{i+1} - x_i)^2= \frac{1}{2}\| x - y \|^2 + \frac{\lambda}{2} \| Dx\|^2$$


 

where $$D = \begin{pmatrix}
   -1 & 1 & 0 & \cdots & 0\\
   0 & -1 & 1 & \cdots & 0\\
   \vdots & \vdots & \ddots & \ddots & 0\\
  0 & 0 & \cdots & -1 & 1\\
   0 & 0 & \cdots & \cdots & 0
  \end{pmatrix} $$

```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
N = 100
y = np.array([0]*N)
x0 =  np.array([0.5]*N)
alpha = 0.01
eps = 0.01
Nmax = 10000
l = 1
mu = 0.01
```

```python
def D_mat(N):
    return np.diag(np.concatenate((np.ones(N-1)*-1,np.array([0])))) + np.diag(np.ones(N-1), 1)

print(D_mat(5))

```

2. f is composed of sums of differential functions so f is differentiable

```python
def f(x):
    D = D_mat(N)
    return 0.5 * np.linalg.norm(x - y)**2 + l * 0.5  *np.linalg.norm(D*x)**2


    
```

```python
def grad_f(x):
    D = D_mat(N)
    return (x-y) + l*np.dot(np.dot(D, np.transpose(D)), x)
   
```

```python
def generate_signal(N,sigma=0.05):
  t = np.linspace(0,1,N)
  t1 = 0.1+0.25*np.random.random()
  t2 = 0.35+0.25*np.random.random()
  yi = np.array([-0.1,0.8,0.2])
  y = np.zeros(N)
  for i in range(y.size):
    if t[i]<=t1:
      y[i]=yi[0]
    elif t[i]>t1 and t[i]<=t2:
      y[i] = yi[1]
    else:
      y[i] = yi[2]
  y += sigma*(2*np.random.random(y.size)-1)
  return t,y
```

```python
t,y = generate_signal(N = 100)
plt.plot(t, y)
plt.show()
```

```python
def gradient_met(grad_f, x0, alpha, eps, Nmax):

  n = 0
  x = x0
  cvg = 1
  norm = eps + 2
  while (n<Nmax and norm > eps):
    #if (n%10 == 0):
     # print(n, norm)
    nx = x - alpha*grad_f(x)
    x = nx
    n = n+1
    norm = np.linalg.norm(grad_f(x))
    

  if (norm > eps):
          cvg =0   
  return x,n, cvg # returns
  # x the minimum,
  # n the number of iterations
  # cvg a boolean that indicates if the algorithm has converged


```

```python
def grad_fmu(x):

  
  g = np.array([x[i] - y[i] for i in range(N)])
  g[0] = g[0] - l*(x[1] - x[0])/(np.sqrt((x[1] - x[0])**2 + mu*mu))

  for i in range(1,N-1):
    a = x[i] - x[i-1]
    b = x[i+1] - x[i]
    g[i] = g[i] + l*(a/np.sqrt(a*a + mu*mu) - b/np.sqrt(b*b + mu*mu))

  g[N-1] = g[N-1] + l*(x[N-1] - x[N-2])/(np.sqrt((x[N-1] - x[N-2])**2 + mu*mu))

  return g
```

```python
def BBstep(grad, x, xm1) :
  dx = x - xm1
  df = grad(x) - grad(xm1)
 
  return np.sum(df * dx)/(np.linalg.norm(df))**2

```

```python
def barzilai_borwein(grad, x0, eps, Nmax):

  n = 0
  xp = -x0
  x = x0
  cvg = 1
  norm = eps + 2
  while (n<Nmax and norm > eps):
    if n%10 == 0:
      print(n, norm )
    alpha = BBstep(grad, x, xp)
    nx = x - alpha*grad(x)
    xp = x
    x = nx
    n = n+1
    norm = np.linalg.norm(grad(x))
    

  if (norm > eps):
          cvg =0   
  return x,n, cvg 
```

```python
def deltax(A):
  dx = np.zeros((Mim, Nim))
  for i in range(Mim):
    for j in range(Nim):
      if i != Mim -1 :
        dx[i,j] = A[i+1,j] - A[i,j];
      else:
        dx[i,j] = 0
  return dx

def deltay(A):
  dy = np.zeros((Mim, Nim))
  for i in range(Mim):
    for j in range(Nim):
      if j != Nim -1 :
        dy[i,j] = A[i,j+1] - A[i,j];
      else:
        dy[i,j] = 0
  return dy
```
